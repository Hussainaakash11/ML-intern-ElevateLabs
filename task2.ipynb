```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setting Seaborn theme
sns.set(style="whitegrid")
```

```python
# You can load from seaborn's built-in dataset or CSV
df = sns.load_dataset('titanic')  # if using seaborn
# OR
# df = pd.read_csv('path_to_titanic.csv')
```

```python
# Understand structure
df.info()

# Basic statistical details
df.describe()

# Unique value counts (for categorical features)
for col in df.select_dtypes(include='object'):
    print(f"{col} value counts:")
    print(df[col].value_counts())
    print("-" * 40)
```

```python
df.hist(figsize=(14, 10), bins=30)
plt.suptitle('Histograms of Numerical Features', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['age'])
plt.title('Boxplot of Age')
plt.show()
```

```python
df['class'].value_counts().plot(kind='bar')
plt.title('Distribution of Passenger Classes')
plt.show()
```

```python
sns.pairplot(df, hue="survived", diag_kind="kde")
plt.suptitle('Pairplot of Features colored by Survival', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="sex", y="age", data=df)
plt.title('Age distribution across Genders')
plt.show()
```

```python
sns.catplot(x="class", hue="sex", col="survived",
            data=df, kind="count", height=4, aspect=.7)
plt.suptitle('Survival Count by Class and Gender', y=1.05)
plt.show()
```
